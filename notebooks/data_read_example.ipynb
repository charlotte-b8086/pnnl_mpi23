```python
import paths
import numpy as np
import h5py
from sdfs.geom_mrst import geom_plot, GeomMRST
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
plt.rc('text', usetex=True)
plt.rc('image', cmap='coolwarm')
```

```python
with h5py.File('../data/mpi23_ens_data.h5','r') as f:
    xi_ens   = f['xi_ens'][:]
    ytms_ens = f['ytms_ens'][:]
    u_ens    = f['u_ens'][:]
    yref     = f['yref'][:]
    ypred    = f['ypred'][:]
    Psi_y    = f['Psi_y'][:]
    ytm      = f['ytm'][()]
    Nens     = f['Nens'][()]
    Nxi      = f['Nxi'][()]
```

```python
geom_filename = '../data/geom_1x.mat'
geom = GeomMRST(geom_filename)
```

```python
fig, ax = plt.subplots()
plot = geom_plot(geom, ytm + yref, ax)
ax.set_aspect('equal')
ax.axis('off')
ax.autoscale(tight=True)
fig.colorbar(plot, ax=ax)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

py = geom_plot(geom, ytm + ytms_ens[0], ax[0])
ax[0].set_aspect('equal')
ax[0].axis('off')
ax[0].autoscale(tight=True)
plt.colorbar(py, ax=ax[0])

py = geom_plot(geom, u_ens[0], ax[1])
ax[1].set_aspect('equal')
ax[1].axis('off')
ax[1].autoscale(tight=True)
plt.colorbar(py, ax=ax[1])

fig.tight_layout()
fig.savefig('ens_0.png', dpi=150) 
```
